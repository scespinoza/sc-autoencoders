```python
import torch
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

from models import VaDE
```

```python
filepath = '../clean_data/GSE70630.txt'
data = pd.read_csv(filepath, sep='\t', index_col=0).T
data.head()
```

```python
data_scaled = MinMaxScaler().fit_transform(data.values)
data_scaled = pd.DataFrame(data_scaled, columns=data.columns, index=data.index)

tumor_ids = LabelEncoder().fit_transform(data.index.map(lambda i: i.split('_')[0]))

t_cells = pd.read_csv('../t-cells.csv')
filter_genes = [gene for gene in data.columns if gene in t_cells['Gene List'].values]
t_counts = (data_scaled[filter_genes] > 0).sum(1)
```

## VaDE Results

```python
def plot_model(name, n, filter_genes=filter_genes):
    """
    Plot latent space coloured by tumor, predicted cluster and gene expression level.
    gene_filter: list of gene names    
    """
    model = VaDE(in_dim=data_scaled.shape[1], n_classes=n)
    model.load_state_dict(torch.load(('weights/' + name + '_trained.pth').format(n)))
    _, _, _, z = model(torch.Tensor(data_scaled.values))
    cluster = model.predict_cluster(torch.Tensor(data_scaled.values)).detach().numpy()
    z_tsne = TSNE().fit_transform(z.detach().numpy())
    fig, ax = plt.subplots(1, 3, figsize=(16, 5))
    ax[0].scatter(z_tsne[:, 0], z_tsne[:, 1], c=tumor_ids, cmap='rainbow', alpha=0.6)
    ax[0].set_title('Tumor IDS')
    ax[1].scatter(z_tsne[:, 0], z_tsne[:, 1], c=cluster, cmap='rainbow', alpha=0.6)
    ax[1].set_title('Predicted Cluster')
    ax[2].scatter(z_tsne[:, 0], z_tsne[:, 1], c=(data[filter_genes] > 0).any(1), cmap='rainbow', alpha=0.6)
    ax[2].set_title('Filter Genes')
    fig.suptitle("n_components = {}".format(n));
```

```python
plot_model('GSE70630', n=6, filter_genes=['CD80', 'CD86'])    
```
